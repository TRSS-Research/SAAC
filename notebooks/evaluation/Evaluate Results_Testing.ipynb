```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import style
import seaborn as sns

%matplotlib inline  
style.use('fivethirtyeight')
sns.set(style='whitegrid',color_codes=True)


respath='../../data/evaluation/processed/'
```

```python
occ_res = pd.read_csv(respath +'Occupation_Results.csv')
print(len(occ_res))
```

```python
sns.boxenplot(data=occ_res, x='a_median', y='gender_detected_val', scale='linear')
```

```python
sns.violinplot(data=occ_res, x='a_median', y='gender_detected_val', scale='count')
```

```python
sns.histplot(occ_res, x='gender_detected_val' , y='a_median', hue='gender_detected_val', legend=False)
```

```python
sns.catplot(data=occ_res, x='a_median', y='gender_detected_val', kind='violin', color='.8')
sns.swarmplot(data=occ_res, x='a_median', y='gender_detected_val', size=1.5)
```

```python
sns.histplot(occ_res, x='a_median', hue='gender_detected_val', element='poly',
    stat='density')
```

```python
sns.histplot(occ_res, x='a_mean', hue='gender_detected_val', element='poly',
    stat='density')
```

```python
sns.histplot(occ_res, x='a_pct90', hue='gender_detected_val', element='poly',
    stat='density')
```

```python
sns.histplot(occ_res, x='a_pct10', hue='gender_detected_val', element='poly',
    stat='density'
)
```

```python
sns.displot(data=occ_res, x="a_median", hue="gender_detected_val", kind="kde")

sns.displot(data=occ_res, x='a_median', col='gender_detected_val')
```

```python
g = sns.JointGrid(data=occ_res, x='a_median', y='a_mean', hue='gender_detected_val')
g.plot(sns.scatterplot, sns.histplot)
```
